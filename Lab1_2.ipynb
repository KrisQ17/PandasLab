```python
#Zad 1.2

from random import randint

def rN():
    return randint(0, 5)

def rND():
    return randint(0, 19)

names = ['Krzysiek', 'Michał', 'Dominik', 'Paweł', 'Konrad', 'Maciej']
surnames = ['Bomba', 'Malinowski', 'Kowalski', 'Kropacz', 'Kanon', 'Misa']
dates = list(randint(1960, 2000) for x in range(20))
jobs = ['Prawnik', 'Dentysta', 'Budowlaniec', 'Informatyk', 'Piekarz', 'Kurier']
cities = ['Kraków', 'Katowice', 'Poznań', 'Warszawa', 'Lublin', 'Gdańsk']
addresses = ['Żelazna 8', 'Karłowicza 2', 'Miłosza 3', 'Obrońców 10', 'Romańska 11', 'Rentgenowska 99']

data = []

for i in range(20):
    record = [names[rN()], surnames[rN()], cities[rN()], dates[rND()], jobs[rN()], addresses[rN()]]
    data.append(record)
    
frame_02 = pd.DataFrame(data)
frame_02.columns = 'Imię', 'Nazwisko', 'Miasto', 'Rok urodzenia', 'Zawód', 'Adres'
frame_02
```

```python
#Zad 1.3
frame_02.to_csv('osoby.csv', index=False)
```

```python
#Zad 3.3 Dataset 1
print(dt_a.groupby(['age', 'sex', 'trestbps'])['chol'].quantile(.25).min())
print(dt_a.groupby(['age', 'sex', 'trestbps'])['thalach'].quantile(.25).min())
print(dt_a.groupby(['age', 'sex', 'trestbps'])['chol'].quantile(.5).max())
print(dt_a.groupby(['age', 'sex', 'trestbps'])['thalach'].quantile(.5).max())
print(dt_a.groupby(['age', 'sex', 'trestbps'])['chol'].quantile(.75).mean())
print(dt_a.groupby(['age', 'sex', 'trestbps'])['thalach'].quantile(.75).mean())
print("=============================")
print(dt_b.groupby(['age', 'sex', 'bmi'])['children'].quantile(.25).min())
print(dt_b.groupby(['age', 'sex', 'bmi'])['charges'].quantile(.25).min())
print(dt_b.groupby(['age', 'sex', 'bmi'])['children'].quantile(.5).max())
print(dt_b.groupby(['age', 'sex', 'bmi'])['charges'].quantile(.5).max())
print(dt_b.groupby(['age', 'sex', 'bmi'])['children'].quantile(.75).mean())
print(dt_b.groupby(['age', 'sex', 'bmi'])['charges'].quantile(.75).mean())
print("=============================")
print(dt_c.groupby(['Platform', 'Year', 'Genre'])['NA_Sales'].quantile(.25).min())
print(dt_c.groupby(['Platform', 'Year', 'Genre'])['EU_Sales'].quantile(.25).min())
print(dt_c.groupby(['Platform', 'Year', 'Genre'])['NA_Sales'].quantile(.5).max())
print(dt_c.groupby(['Platform', 'Year', 'Genre'])['EU_Sales'].quantile(.5).max())
print(dt_c.groupby(['Platform', 'Year', 'Genre'])['NA_Sales'].quantile(.75).mean())
print(dt_c.groupby(['Platform', 'Year', 'Genre'])['EU_Sales'].quantile(.75).mean())
```
