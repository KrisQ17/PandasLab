2.	Wykonać wybrane 2 z 3 poniższych wykresów kołowych:
- ilości pracowników mieszkających w poszczególnych regionach,
- ilości kobiet mieszkających w poszczególnych hrabstwach,
- ilości mężczyzn z emailami na poszczególnych domenach


```python
import matplotlib.pyplot as plt

path3 = 'https://raw.githubusercontent.com/palles77/Pandas20212022_Lab3/main/employees10000.csv'

frame4 = pd.read_csv(path3)
frame4
```

```python
# Zadanie 1
salary_min = frame4[frame4['Gender']=='M']['Salary'].min()
salary_max = frame4[frame4['Gender']=='M']['Salary'].max()
ranged = tuple(x for x in range(salary_min, salary_max, 50))
frame_male = frame4[frame4['Gender']=='M']
frame_male.groupby(pd.cut(frame_male['Salary'], ranged))['Salary'].size().plot()
```

```python
# Zadanie 1
salary_min = frame4[frame4['Gender']=='F']['Salary'].min()
salary_max = frame4[frame4['Gender']=='F']['Salary'].max()
ranged = tuple(x for x in range(salary_min, salary_max, 50))
frame_female = frame4[frame4['Gender']=='F']
frame_female.groupby(pd.cut(frame_female['Salary'], ranged))['Salary'].size().plot()
```

```python
#Zadanie 2
regions = frame4.groupby(['Region']).size()

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = regions.keys()
sizes = regions.values
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
fig1.set_facecolor('white')
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
#Zadanie 2
counties = frame4.groupby(['County']).size()

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = counties.keys()
sizes = counties.values

fig1, ax1 = plt.subplots()
fig1.set_facecolor('white')
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
